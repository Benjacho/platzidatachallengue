```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.grid()
%matplotlib inline
```

```python
test_df = pd.read_csv('test.csv')
test_df
```

```python
train_df = pd.read_csv('train.csv')
train_df
```

```python
gs_df = pd.read_csv('gender_submission.csv')
gs_df
```

```python
#Merging
df_merged = pd.merge(gs_df, test_df, on='PassengerId')
df_merged
```

```python
df_joined = pd.concat([train_df, df_merged])
df_joined
```

```python
df_joined.set_index('PassengerId', inplace=True)
df_joined
```

```python
rows, columns = df_joined.shape
print(f'en el titanic iban {rows} personas')
```

```python
df_joined[df_joined['Survived'] == 1]['Sex'].value_counts().plot.pie(autopct='%1.1f%%')
```

```python
df_joined[df_joined['Survived'] == 1]['Sex'].value_counts()
```

```python
df_joined[df_joined['Survived'] == 0]['Sex'].value_counts(dropna=False)
```

```python
top_10_survivors_ages = df_joined[df_joined['Survived'] == 1]['Age'].value_counts().head(10).sort_values()
```

```python
top_10_not_survivors_ages = df_joined[df_joined['Survived'] == 0]['Age'].value_counts().head(10).sort_values()
```

```python
fix, ax = plt.subplots()
for i, v in enumerate(top_10_survivors_ages.values):
    ax.text(v, i - .2, str(v), color='black', fontweight='bold')
ax.set_xlabel('Numero de Sobrevivientes')
ax.set_ylabel('Edad')
top_10_survivors_ages.plot(kind='barh')
```

```python
fix, ax = plt.subplots()
for i, v in enumerate(top_10_not_survivors_ages.values):
    ax.text(v, i - .2, str(v), color='black', fontweight='bold')
ax.set_xlabel('Numero de No Sobrevivientes')
ax.set_ylabel('Edad')
top_10_not_survivors_ages.plot(kind='barh')
```

```python
conversion_in_1912 = 4.87
amount_in_usd = df_joined['Fare'].sum() * conversion_in_1912
amount_in_usd
```

This amount is in dollars in 1912, if we convert applying inflation rate it is equal to according to this https://www.in2013dollars.com/us/inflation/1912?amount=1.


```python
inflation_rate = 26.71
dollars_today = amount_in_usd * inflation_rate
dollars_today
```
